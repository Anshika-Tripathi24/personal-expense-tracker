```python
import pandas as pd
df=pd.read_csv('expenses.csv')
df
```

```python
df.info()
```

```python
#Basic analysis
#Total spent 
print('Total amount:',df['Amount'].sum())
```

```python
#Spend by category
print(df.groupby("Category")['Amount'].sum())
```

```python
#Daily spend
print(df.groupby("Date")['Amount'].sum())
```

```python
#Visualization(Matplotlib + Seaborn)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Spend by category(pie chart)
df.groupby('Category')['Amount'].sum().plot.pie(autopct='%1.1f%%')
plt.title('Expense distribution by category')
plt.ylabel(" ")
plt.show()
```

```python
#Daily expenses(line chart)
daily_expense = df.groupby("Date")['Amount'].sum()
daily_expense.plot(linestyle='solid', marker='o',color='red')
plt.title("Daily Expenses")
plt.xlabel("Date")
plt.ylabel("Amount Spent")
plt.grid()
plt.xticks(rotation=0)
#plt.tight_layout()
plt.show()
```

```python
#Category-wise trend(bar plot)
sns.barplot(x='Category',y='Amount',data=df,estimator='sum',palette='viridis')
plt.title('Total spend by Category')
#plt.tight_layout()
plt.show()
```

```python
#highest spending category
print("Max Spend Category:", df.groupby("Category")['Amount'].sum().idxmax())
```

```python
#save reports
df.groupby("Category")['Amount'].sum().to_csv("category_summary.csv")
```
